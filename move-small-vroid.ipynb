```python
from glob import glob
import os
import shutil

from PIL import Image
from tqdm import tqdm
```

```python
datadir = 'datasets/vroid-images'
errdir = 'datasets/v-errors'
minsize = 256

os.makedirs(errdir, exist_ok=True)

files = glob(os.path.join(datadir, '*'), recursive=True)
```

```python
def dmove(src, dst):
    if os.path.isdir(src):
        print(f"src is dir {src}")
        return
    dir = os.path.dirname(src)
    distdir = os.path.join(dst, dir)
    print(f"mkdir {distdir}")
    os.makedirs(distdir, exist_ok=True)
    try:
        print(f"move {src} to {distdir}")
        shutil.move(src, distdir)
    except:
        pass

for fname in tqdm(files):
    try:
        image = Image.open(fname)
        w, h = image.size
        if w < minsize or h < minsize:
            print(f"small file: {fname}")
            dmove(fname, errdir)
    except Exception as e:
        print(f"error file: {fname}, {e}")
        dmove(fname, errdir)
```
